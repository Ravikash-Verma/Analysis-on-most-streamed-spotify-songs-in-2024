# Top 10 Artists based on spotify streams

```python
top10=df.sort_values(by='Spotify Streams',ascending=False).head(10)
```

```python
top10=top10.sort_values(by='Spotify Streams',ascending=True)
```

```python
top10.reset_index()
```

```python
plt.bar(top10['Artist'],top10['Spotify Streams'] , color='skyblue')
plt.xticks(rotation=45)
plt.xlabel('Artists')
plt.ylabel('Spotify Streams')
plt.title('Top 10 Artists based on spotify streams')
```
